```python
import json

data = []

with open('data/locallikes_obs_atom.txt') as file:
    for line in file:
        line = line.strip()
        val = line.split(', ')
        x_val = val[0].split(': ')[1]
        y_val = val[1].split(': ')[1]
        data.append({'x': int(x_val), 'y': int(y_val)})

print(json.dumps(data, indent = 4))
```
